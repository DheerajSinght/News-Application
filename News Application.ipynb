```python
import io
import requests
import webbrowser
from tkinter import *
from PIL import ImageTk,Image
from urllib.request import urlopen


class NewsApp():
    
    def __init__(self):
        
        self.data=requests.get('https://newsapi.org/v2/top-headlines?country=in&apiKey=d25625f3fd9c4a08b32801c9d7a2848d').json()
        self.load_gui()
        self.load_news_item(0)
    
    def load_gui(self):
        self.root=Tk()
        self.root.geometry('350x600')
        self.root.title('My News')
        self.root.resizable(0,0)
        self.root.configure(background='black')
        self.root.iconbitmap('photo.ico')
        
    def clear(self):
        for i in self.root.pack_slaves():
            i.destroy()
        
        
    def load_news_item(self,index):
        
        
        self.clear()
        # load image
        try:
            img_url=self.data['articles'][index]['urlToImage']
            raw_data=urlopen(img_url).read()
            im=Image.open(io.BytesIO(raw_data)).resize((350,220))
            photo=ImageTk.PhotoImage(im)
            
        except:            
            img=Image.open('download.png')
            resize_img=img.resize((350,220))
            photo=ImageTk.PhotoImage(resize_img)
            
            
        label_img=Label(self.root,image=photo)
        label_img.pack()
        
        
        frame = Frame(self.root, bg='black')
        frame.pack(expand=True, fill=BOTH)

        # Create a frame for the labels
        labels_frame = Frame(frame, bg='black')
        labels_frame.pack(side=TOP, fill=BOTH, expand=True)

        # Heading Label
        Heading = Label(labels_frame, text=self.data['articles'][index]['title'], bg='black', fg='white', justify='center', wraplength=350)
        Heading.pack(pady=(10, 20))
        Heading.configure(font=('verdana', 15))

        # Details Label
        details = Label(labels_frame, text=self.data['articles'][index]['description'], bg='black', fg='white', wraplength=350, justify='center')
        details.pack(pady=(1, 20))
        details.config(font=('verdana', 12))

        # Buttons in the same frame
        if index != 0:
            prev_bttn = Button(frame, text='Prev', bg='white', fg='black', width=16, height=3, command=lambda: self.load_news_item(index - 1))
            prev_bttn.pack(side=LEFT)

        read_bttn = Button(frame, text='Read More', width=16, height=3, bg='white', fg='black', command=lambda: self.open_link(self.data['articles'][index]['url']))
        read_bttn.pack(side=LEFT)

        if index != len(self.data['articles']) - 1:
            next_bttn = Button(frame, text='Next', width=16, height=3, bg='white', fg='black', command=lambda: self.load_news_item(index + 1))
            next_bttn.pack(side=LEFT)    
        self.root.mainloop()
    
    def open_link(self,url):
        webbrowser.open(url)
    


obj=NewsApp()
```
